```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([500, 750, 1000, 1250, 1500, 1750, 2000]).reshape(-1, 1)
y = np.array([150, 200, 250, 275, 300, 325, 350])  # price in $1000s
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
print("Slope (price per sqft):", model.coef_[0])
print("Intercept (base price):", model.intercept_)
```

```python
new_sizes = np.array([[2200], [2500]])
new_prices = model.predict(new_sizes)
print("Predicted prices for new house sizes:", new_prices)
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.scatter(new_sizes, new_prices, color='green', marker='x', label='Predicted Prices')
plt.xlabel('House Size (sqft)')
plt.ylabel('House Price ($1000s)')
plt.title('House Price Prediction with Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
